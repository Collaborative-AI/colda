```python
all_labels = list(bat["Name"].unique())
all_labels[:5]
```

```python
def get_csv(label):
    temp = yf.Ticker(label)
    df_temp = temp.history(period="max")
    return df_temp.to_csv(label + ".csv")
    # print(df_temp[1:])
    return df_temp.head()
```

```python
for i in all_labels:
    # print(i)
    get_csv(i)
```
